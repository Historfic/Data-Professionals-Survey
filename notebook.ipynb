```python
# Define a list of valid job titles
valid_job = ['data scientist', 'data engineer', 'data analyst', 'database developer', 'data architect', 'others']

# Convert column values to lowercase for uniformity
df['Q1 - Which Title Best Fits your Current Role?'] = df['Q1 - Which Title Best Fits your Current Role?'].str.lower()

# Function to map values to standard job titles
def clean_job_title(title):
    if 'data scientist' in title:
        return 'data scientist'
    elif 'data engineer' in title:
        return 'data engineer'
    elif 'data analyst' in title or 'analyst' in title:
        return 'data analyst'
    elif 'database developer' in title:
        return 'database developer'
    elif 'data architect' in title:
        return 'data architect'
    else:
        return 'others'

# Apply the cleaning function to the column
df['Q1 - Which Title Best Fits your Current Role?'] = df['Q1 - Which Title Best Fits your Current Role?'].apply(clean_job_title)

# Count occurrences of each job title
job_title_counts = df['Q1 - Which Title Best Fits your Current Role?'].value_counts()

# Display the counts
print(job_title_counts)
```

### Cleaning Question 3

```python
# Function to calculate the midpoint of a salary range
def calculate_midpoint(salary_range):
    # Remove the 'k' character
    salary_range = salary_range.replace('k', '')
    # Split the range into lower and upper bounds
    lower, upper = map(int, salary_range.split('-'))
    # Calculate the midpoint
    midpoint = (lower + upper) / 2
    return midpoint

# Apply the function to the column
df['Salary Midpoint'] = df['Q3 - Current Yearly Salary (in USD)'].apply(calculate_midpoint)

# Calculate the average salary midpoint
average_salary = df['Salary Midpoint'].mean()

# Display the average salary
print(f"The average salary is: {average_salary:.2f}")
```

### Cleaning Question 5

```python
df['Q5 - Favorite Programming Language'].str.lower()
print(df['Q5 - Favorite Programming Language'].unique())
```

```python
valid_languages = ['python', 'r', 'sql', 'java', 'javascript']

df['Q5 - Favorite Programming Language'] = df['Q5 - Favorite Programming Language'].str.lower()

def clean_language(language):
    if language in valid_languages:
        return language
    elif 'sql' in language:
        return 'sql'
    elif 'java' in language:
        return 'java'
    elif 'javascript' in language or 'js' in language:
        return 'javascript'
    elif 'r' in language:
        return 'r'
    else:
        return 'others'
    
df['Q5 - Favorite Programming Language'] = df['Q5 - Favorite Programming Language'].apply(clean_language)
df.dropna(subset=['Q5 - Favorite Programming Language'], inplace=True)
print(df['Q5 - Favorite Programming Language'].unique())

```
